```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import cv2
```

```python
def gaussian_blur_image(img, k: int = 5):
    blurred = cv2.GaussianBlur(img, (k, k), 0)    
    return blurred

def show_image(img, title: str = "Image"):
    img_corrected = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_corrected)
    plt.axis("off")
    plt.title(title)
    plt.show()
```

```python
image_path = Path("./images/000010.jpg")

img = cv2.imread(str(image_path), cv2.IMREAD_UNCHANGED)
blurred_image = gaussian_blur_image(img, k=9)
```

```python
show_image(img, title="Original Image")
show_image(blurred_image, title="Blurred Image")
```

```python
out_path = image_path.with_name(image_path.stem + "_blurred" + image_path.suffix)

# Use same image format as input
success = cv2.imwrite(str(out_path), blurred_image)
```

```python
# generate random image
h = 218
w = 178
img = np.random.randint(0, 256, (h, w, 3), dtype=np.uint8)
random_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

show_image(random_img, title="Random Image")
```

```python
cv2.imwrite("./images/random_image.jpg", random_img)
```
